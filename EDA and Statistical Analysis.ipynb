###### Outlier Detection:

```python
# Calculate quartiles and IQR for price
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers based on IQR method
outliers_iqr = df[(df['price'] < Q1 - 1.5 * IQR) | (df['price'] > Q3 + 1.5 * IQR)]
print("Outliers based on IQR method:")
print(outliers_iqr['price'])
```

###### 1. Distribution of Listing Prices across Different Neighborhoods

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

# Visualization: Box plot of prices by neighborhood
plt.figure(figsize=(12, 8))
sns.boxplot(x='neighbourhood_group', y='price', data=df)
plt.title('Price Distribution across Neighborhoods')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.show()

# Statistical Test: ANOVA
f_statistic, p_value = f_oneway(
    df[df['neighbourhood_group'] == 'Manhattan']['price'],
    df[df['neighbourhood_group'] == 'Brooklyn']['price'],
    df[df['neighbourhood_group'] == 'Queens']['price'],
    df[df['neighbourhood_group'] == 'Staten Island']['price'],
    df[df['neighbourhood_group'] == 'Bronx']['price']
)

print(f"ANOVA Results:")
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

###### 2. Comparison of Room Types and Prices

```python
from scipy.stats import ttest_ind, ranksums

# Visualization: Box plot of prices by room type
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price (USD)')
plt.show()

# Statistical Test: T-test or Wilcoxon rank-sum test
entire_home_prices = df[df['room_type'] == 'Entire home/apt']['price']
private_room_prices = df[df['room_type'] == 'Private room']['price']
shared_room_prices = df[df['room_type'] == 'Shared room']['price']

t_statistic, p_value = ttest_ind(entire_home_prices, private_room_prices)
# Or ranksums for non-parametric test
# u_statistic, p_value = ranksums(entire_home_prices, private_room_prices)

print(f"T-test Results for Entire Home vs Private Room:")
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```

###### 3. Relationship Between Number of Reviews and Price

```python
# Visualization: Scatter plot of number of reviews vs price
plt.figure(figsize=(10, 6))
sns.regplot(x='number_of_reviews', y='price', data=df)
plt.title('Number of Reviews vs Price')
plt.xlabel('Number of Reviews')
plt.ylabel('Price (USD)')
plt.show()

# Statistical Test: Pearson correlation
correlation = df[['number_of_reviews', 'price']].corr().iloc[0, 1]
print(f"Pearson Correlation Coefficient: {correlation}")
```

#####  Hypothesis Test for Difference in Mean Prices between Neighborhoods

```python
from scipy.stats import f_oneway

# Example: Testing mean prices between Manhattan, Brooklyn, and Queens
manhattan_prices = df[df['neighbourhood_group'] == 'Manhattan']['price']
brooklyn_prices = df[df['neighbourhood_group'] == 'Brooklyn']['price']
queens_prices = df[df['neighbourhood_group'] == 'Queens']['price']

f_statistic, p_value = f_oneway(manhattan_prices, brooklyn_prices, queens_prices)

print(f"ANOVA Results:")
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```
